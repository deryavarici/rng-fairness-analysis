```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# --- ADIM 1: VERİ ÜRETİMİ ---
oyuncu_verileri = []
# 10.000 oyuncu için %5 şansla kılıç düşme simülasyonu
for oyuncu_id in range(1, 10001):
    kilic_dustu_mu = 1 if random.random() < 0.05 else 0
    oyuncu_verileri.append([oyuncu_id, kilic_dustu_mu])

# Veriyi tabloya (DataFrame) dönüştürme
df = pd.DataFrame(oyuncu_verileri, columns=['Oyuncu_ID', 'Kilic_Dustu'])

# --- ADIM 2: ANALİZ VE MATEMATİKSEL KONTROL ---
toplam_kilic = df['Kilic_Dustu'].sum()
gerceklesen_oran = (toplam_kilic / 10000) * 100

print(f"Teorik Beklenen Oran: %5.00")
print(f"Gerçekleşen Düşürme Oranı: %{gerceklesen_oran:.2f}")
print("-" * 40)

# Otomatik Sonuç Çıktısı
if 4.9 <= gerceklesen_oran <= 5.1:
    print("✅ ANALİZ SONUCU: Sistem adil! 10.000 deneme sonucunda gerçekleşen oran, beklenen %5'lik hedefe istatistiksel olarak ulaşmıştır. RNG algoritması doğru çalışıyor.")
else:
    print("⚠️ ANALİZ SONUCU: Sistemde sapma var! Beklenen orandan uzaklaşıldı.")

# --- ADIM 3: GÖRSELLEŞTİRME ---
etiketler = ['Kılıç Düşürenler', 'Düşüremeyenler']
sayilar = [toplam_kilic, 10000 - toplam_kilic]

plt.figure(figsize=(6, 6))
plt.pie(sayilar, labels=etiketler, autopct='%1.2f%%', colors=['#FFD700', '#E0E0E0'], startangle=90)
plt.title(f"10.000 Oyuncu: Efsanevi Eşya Düşürme Analizi\n(Gerçekleşen: %{gerceklesen_oran:.2f} | Beklenen: %5.00)", fontsize=12)
plt.show()
```
